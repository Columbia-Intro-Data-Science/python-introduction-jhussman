## Motivation

The energy industry is poised for enormous disruption in the coming years and decades. Renewable energies are increasingly cost-competitive with traditional fossil fuel powered generation, and new technologies for energy efficiency are challenging traditional generation as well. A key factor for the future transformation of energy systems around the world relies on management of energy demand.

Energy demand is an inherently stochastic process. Peaks in energy demand are especially important for the following reasons:

1. Renewable energies such as wind and solar are intermittent by nature; solar panels only generate power when the sun is shining, and wind turbines only when the wind is blowing. Reducing peaks in demand helps balance the intermittent generation from these sources, particularly when the timing and magniute of demand does not coincide with the renewable generation. Excess energy from renewables can be stored in batteries, for use during peak demand. Lowering peaks in demand lowers the capacity of energy storage required, and encourages a shift towards renewables.

```python
from IPython.display import Image
Image("../Peak-shaving-and-valley-filling.jpg")
```

2. Currently, fossil fuel power plants are relied on to provide "backup" capacity to meet peak demand. In many cases, fossil fuel plants are paid to operate in order to meet a peak demand level which may never appear. Reducing peak demand would lower the necessary capacity of fossil fuel plants, an important step in addressing climate change and discouraging investment in fossil fuel generation.

3. When demand peaks, the potential for power outages is at its greatest, as utilities may not have enough capacity to cover peak demand. Reducing peaks in demand increases the security of customers and reduces risks of blackouts.

An important strategy for addressing peak demand is known as "Demand Response" (DR). DR refers to programs in which customers agree to have their power supply cut at times of peak demand, in exchange for payment by the utility. These customers are presumably equipped with backup energy storage, reduction, or generation capabilities. The utility benefits by reducing their total power requirement, and risk of supply shortages. Other customers benefit from lower prices.

## Project Description

We propose a solution to help identify candidates for DR, and recommend DR characterstics, based on electric load time series data. Our solution will look for patterns in electric demand and classify a number of demand profiles, which are sometimes referred to as "energy personalities". 

```python
Image("../load-curves_600px.png")
```

Our solution is targeted at utilities, renewable or distributed generation companies who wish to indentify candidates for DR. Users will interact with the solution by inputting an electrical load time series dataset, which may be residential or commerical. The solution will classify the demand data according to its "energy personality". It will then recommend whether this load series is a good candidate for a DR program.
