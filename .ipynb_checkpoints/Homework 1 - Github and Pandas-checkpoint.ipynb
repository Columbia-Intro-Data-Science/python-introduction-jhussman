## How could you have improved performance?

*Hint:* Try plotting the data in three dimensions along with the hyperplane solution, and see if you can infer
a new variable which will help, or try a nonlinear/non-parametric model

ANSWER:
Newspaper appeared to lead to the least amount of sales (basically zero), so it would be useful to divert newspaper sales to that of TV and radio to improve performance

````python
import statsmodels.formula.api as sm
from mpl_toolkits.mplot3d import Axes3D

tv_radio = df[['TV','Radio','Sales']]

model = sm.ols(formula='Sales ~ TV + Radio', data = tv_radio)
fit = model.fit()

fit.summary()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x_surf = numpy.arange(0, 350, 20)                # generate a mesh
y_surf = numpy.arange(0, 60, 4)

x_surf, y_surf = numpy.meshgrid(x_surf, y_surf)

exog = pd.core.frame.DataFrame({'TV': x_surf.ravel(), 'Radio': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.reshape(x_surf.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.4)

ax.scatter(df['TV'], df['Radio'], df['Sales'],
           c='blue',
           marker='o',
           alpha=1)

ax.set_xlabel('TV')
ax.set_ylabel('Radio')
ax.set_zlabel('Sales')
````

## Problem 3:  Gradient Descent and the learning rate

By modifying the learning rate below, show how the convergence takes longer or doesn't converge at all.
Can you explain in words or math why this is?

ANSWER:
As the learning rate is increased, convergence occurs faster because the algorithm descends down the gradient at a larger rate every iteration. However, if the learning rate is too big, it won't converge as it will overshoot the miniminum.  

With the initial learning rate of LR = 0.0001: b = 0.0351, m = 1.479.

When the learning rate is changed to LR = 0.001 (10x initial), b = -3.12 and m = -1.58. This means that they don't converge. When the learning rate is change to LR = 0.00001 (0.1x initial) b = 0.0294, m = 1.47. This means that it hasn't yet converged due to the smaller size of the learning rate. 

````python
from numpy import *
import pylab as plt

# y = mx + b
# m is slope, b is y-intercept
def compute_error_for_line_given_points(b, m, points):
    totalError = 0
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        totalError += (y - (m * x + b)) ** 2
    return totalError / float(len(points))

def step_gradient(b_current, m_current, points, learningRate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for i in range(0, len(points)):
        x = points[i, 0]
        y = points[i, 1]
        b_gradient += -(2/N) * (y - ((m_current * x) + b_current))
        m_gradient += -(2/N) * x * (y - ((m_current * x) + b_current))
    new_b = b_current - (learningRate * b_gradient)
    new_m = m_current - (learningRate * m_gradient)
    return [new_b, new_m]

def gradient_descent_runner(points, starting_b, starting_m, learning_rate, num_iterations):
    b = starting_b
    m = starting_m
    for i in range(num_iterations):
        b, m = step_gradient(b, m, array(points), learning_rate)
    return [b, m]

def run(num_iterations):
    points = genfromtxt("../python-introduction-jhussman/data.csv", delimiter=",")
    learning_rate = 0.0001
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    print("Starting gradient descent at b = {0}, m = {1}, error = {2}".format(initial_b, initial_m, compute_error_for_line_given_points(initial_b, initial_m, points)))
    print("Running...")
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    print("After {0} iterations b = {1}, m = {2}, error = {3}".format(num_iterations, b, m, compute_error_for_line_given_points(b, m, points)))
    for i in range(0,len(points)):
        plt.scatter(points[i,0],points[i,1])
        plt.scatter(points[i,0],m*points[i,0]+b,color='r')
    return [compute_error_for_line_given_points(b, m, points)]
run(100)
````

````python
# Use this to visually discuss convergence rate based on learning rate

for num in range(0,10):
    run(num)
    plt.show()

#The first plot shows the horizontal red line as the initial guess of all the points and the fitted red line as the
#converged solution after 10 iterations. The proceeding plots (2-10) show the subsequent iterations where the red line 
#moves towards fitting the data. Plots 1-6 best visualize this. After plot 6, the convergence is too small to see.
````

## Problem 3 Part 2

Plot the error as a function of the number of iterations for various learning rates. Choose the rates
so that it tells a story.



        

````python
def err(num_iterations, learning_rate):
    points = genfromtxt("../python-introduction-jhussman/data.csv", delimiter=",")
    initial_b = 0 # initial y-intercept guess
    initial_m = 0 # initial slope guess
    num_iterations = num_iterations
    [b, m] = gradient_descent_runner(points, initial_b, initial_m, learning_rate, num_iterations)
    return [compute_error_for_line_given_points(b, m, points)]


initial = 0.000050
initial2 = 2*initial
initial3 = 1.5*initial2
initial4 = 4*initial
error = []
error2 = []
error3 = []
error4 = []
iterate = []

for num in range(0,11):
    error.append(err(num,initial)) 
    error2.append(err(num,initial2))
    error3.append(err(num,initial3))
    error4.append(err(num,initial4))
    iterate.append(num)
    
plt.plot(iterate, error, label='Learning Rate = %s' %initial)
plt.plot(iterate, error2,label='Learning Rate = %s' %initial2)
plt.plot(iterate, error3,label='Learning Rate = %s' %initial3)
plt.plot(iterate, error4,label='Learning Rate = %s' %initial4)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.title('Error vs. Iteration for various learning rates')
plt.show()  
````
